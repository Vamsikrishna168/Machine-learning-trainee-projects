```python
fake['genuiness']=0
genuine['genuiness']=1
```

```python
dataset=pd.concat([fake,genuine])
dataset
```

## PRE PROCESSING

```python
dataset=dataset.reset_index(drop=True)
```

```python
dataset
```

```python
dataset=dataset.drop(['date','title','subject'],axis=1)
```

```python
dataset
```

```python
from nltk.tokenize import word_tokenize
dataset['text']=dataset['text'].apply(word_tokenize)
```

```python
dataset
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
sb=SnowballStemmer('english')
```

```python
def stem_it(text):
    return [sb.stem(word) for word in text] 
```

```python
dataset['text']=dataset['text'].apply(stem_it)
```

```python
dataset
```

```python
def stopword_revomal(text):
    return [word for word in text if len(word) >> 2]
```

```python
dataset['text']=dataset['text'].apply(stopword_revomal)
```

```python
dataset
```

```python
dataset['text']=dataset['text'].apply(' '.join)
dataset
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(dataset['text'],dataset['genuiness'],test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
Tdidfv=TfidfVectorizer(max_df=0.7)
```

```python
Tdidfv_train=Tdidfv.fit_transform(x_train)
Tdidfv_test=Tdidfv.transform(x_test)
```

```python
print(Tdidfv_train)
```

```python
print(Tdidfv_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(max_iter=500)
lr.fit(Tdidfv_train,y_train)
```

```python
y_pred=lr.predict(Tdidfv_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier

pac=PassiveAggressiveClassifier(max_iter =100)
pac.fit(Tdidfv_train,y_train)
```

```python
y_pred_pac=pac.predict(Tdidfv_test)
accuracy_score(y_test,y_pred_pac)
```
